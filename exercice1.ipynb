```python
def multiplication(n1, n2):
    resultat = 0
    if n1 >= 0:
        for i in range(n1):
            resultat += n2
        return resultat
    else:
        for i in range(abs(n1)):
            resultat = resultat - n2
        return resultat
```

```python
multiplication(3, 5)
```

```python
multiplication(-4, -8)
```

```python
multiplication(-2, 6)
```
